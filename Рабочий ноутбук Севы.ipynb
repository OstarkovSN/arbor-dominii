```python
df_company, df_founder_legal, df_founder_natural = get_dfs()

display(df_founder_legal.head(20))
```

```python
# Группируем по ФИО и подсчитываем количество уникальных 'inn'
fio_inn_counts = df_founder_natural.groupby(
    ['last_name', 'first_name', 'second_name']
)['inn'].nunique().reset_index(name='unique_inn_count')

# Вывод промежуточного результата
print("\nКоличество уникальных 'inn' для каждого ФИО:")
print(fio_inn_counts)
```

```python
# Находим индекс строки с максимальным количеством уникальных 'inn'
max_idx = fio_inn_counts['unique_inn_count'].idxmax()

# Извлекаем соответствующее ФИО и количество
top_fio = fio_inn_counts.loc[max_idx]

print("\nФИО с наибольшим количеством уникальных 'inn':")
print(f"Фамилия: {top_fio['last_name']}, Имя: {top_fio['first_name']}, Отчество: {top_fio['second_name']}, Количество уникальных 'inn': {top_fio['unique_inn_count']}")
```
